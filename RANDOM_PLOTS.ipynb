```python
import random
import numpy as np
from matplotlib import pyplot as plt
i = 0
x = np.random.randint(1, 1001, 1000)
y = np.random.randint(1, 1001, 1000)
a = np.random.randint(1, 101, 1000)
r = np.random.randint(0, 2, 1000)
g = np.random.randint(0, 2, 1000)
b = np.random.randint(0, 2, 1000)
while i <1000:
  rbg = (r[i], b[i], g[i])
  plt.scatter(x[i], y[i], s=a[i], color=rbg)
  i = i + 1
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("SCATTER PLOT")
plt.show()
```
